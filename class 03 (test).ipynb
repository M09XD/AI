### Question 1

```python
import numpy as np

def reshape2d_to_1d(arr):
    m = arr.reshape(-1)
    return [m]

y = np.array([[1, 2], [3, 4]])
print(y)

a = reshape2d_to_1d(y)
print(a)
```

### Question 2

```python
n = int(input())
sum = 0
if n%2!=0:
  i = 1
  while i <= n:
     sum += i**2
     i += 1
  print(sum)
elif n%2==0:
  i = 1
  while i <= n:
     sum += (i+1)**2
     i += 1
  print(sum)
else: print("ERROR") 

```

### Question 3

```python
lst = [1, 2, 3, 4, 6]

def reverse_inner_list(lst):
    if len(lst) >= 3:
        return lst[1:-1][::-1]
    return "ERROR"

b = reverse_inner_list(lst)
print(b)
```
